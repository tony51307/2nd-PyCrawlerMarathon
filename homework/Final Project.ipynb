## 主題：爬取Cupoy官網熱門新聞

透過觀察Cupoy新聞官網可以發現，Cupoy官網為動態型網頁，所以此次採用了Selenium及BeautifulSoup來爬取該網站文章。

### Step 1 導入需要的套件

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import time
```

### Step 2 開始針對網站內容爬取

```python
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get("https://www.cupoy.com/newsfeed/topstory")
time.sleep(20) # 因網速較慢，所以等待20秒來讓網站完全讀取

title=[]
source=[]
website_link=[]

# 自動將網頁往下滑，來更新文章
for i in range(50):
    browser.execute_script("window.scrollTo(0, 1000000);") 
    time.sleep(5) 
    soup=BeautifulSoup(browser.page_source)
    article_title=soup.find(class_='ReactVirtualized__Grid__innerScrollContainer').findAll('div',class_='sc-eEieub sc-iuDHTM emMJDZ',recursive=False)
    # 將爬到的文章標題、來源、連結記錄起來
    for t in article_title:
        title.append(t.find(class_='sc-erNlkL sc-ekulBa gyyrZF').text)
        source.append(t.find(class_='sc-gacfCG bPSpUf').text)
        website_link.append(t.findAll('a')[0].get('href'))
```

```python
# 將資料儲存成DataFrame
import pandas as pd
df=pd.DataFrame({'Article Title':title, 'Article Source':source, "Article Link":website_link})
```

### Step 3 產生統計圖表

```python
#導入圖表套件
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 統計不同來源的文章數
sns.set('poster',font=['sans-serif'])
sns.set_style({"font.sans-serif":['Microsoft JhengHei']}) # 使seaborn可以顯示中文
sns.catplot(x='Article Source',data=df,height=10,aspect=5,kind='count')
plt.xlabel('新聞來源',size=50)
plt.xticks(rotation=90,size=50)
plt.yticks(size=50)
plt.ylabel('數量',size=50)
plt.show()
```

從以上圖表可以看出，最近育兒、足球等相關文章近期受到許多關注，而職涯、科技類的文章跟平常一樣都有受到一定程度的關注。

### Step 4 針對標題文字分析

```python
# 導入jieba套件來斷詞
import jieba
import jieba.analyse

title_merged=' '.join(list(df['Article Title']))

# 移除特殊符號
for i in '，。：！! ?？|「」、-《》｜／':
    title_merged=title_merged.replace(i,'')
    title_merged.lower()

# 因第一次斷詞後發現有些詞被錯誤切割，所以自行增加詞庫來使斷詞更加地準確
jieba.load_userdict('userDict.txt')

extract_words=jieba.lcut(title_merged)

# 導入停用詞列表
stopwords=[]
with open('停用詞-繁體中文.txt', 'r') as f:
    for line in f.readlines():
        stopwords.append(line.replace('\n',''))

# 移除斷詞列表中的停用詞
extract_words=list(filter(lambda x : x not in stopwords, extract_words))
```

```python
# 統計詞頻
import collections
word_frequency=collections.Counter(extract_words)
word_frequency
```

```python
# 產生文字雲來更好視覺化
from wordcloud import WordCloud
wordcloud = WordCloud(font_path="/System/Library/fonts/PingFang.ttc", background_color='white')
wordcloud.generate_from_frequencies(frequencies=word_frequency)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

從以上結果可以看出，絕大多熱門文章的標題主要都是關於旅遊、球迷和科技。
