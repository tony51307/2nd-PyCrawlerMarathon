### ① 取出今天所有的發文

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
r=requests.get('https://www.ettoday.net/news/news-list.htm')
soup=BeautifulSoup(r.text)
for d in soup.find(class_="part_list_2").find_all('h3'):
    if pd.to_datetime(d.find(class_="date").text)>pd.to_datetime("March 08 2020 23:59:59"):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df=pd.DataFrame()
Date=[]
Type=[]
Topic=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    if pd.to_datetime(d.find(class_="date").text)>pd.to_datetime("March 08 2020 23:59:59"):
        Date.append(pd.to_datetime(d.find(class_='date').text))
        Type.append(d.find_all('em')[-1].text)
        Topic.append(d.find_all('a')[-1].text)
df['date']=Date
df['type']=Type
df['topic']=Topic

print(df)
```

### ③ 哪一個類別的文章最多

```python
print(df['type'].value_counts())
print('The type with the highest number of news: '+str(df['type'].value_counts().index[0]))
```
