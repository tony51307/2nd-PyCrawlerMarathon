## 作業目標

1. 取出今天所有的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
from datetime import date
import pandas as pd


browser = webdriver.Chrome(ChromeDriverManager().install())

browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(40):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 1000000);")

source=browser.page_source
soup=BeautifulSoup(source)

Date=[]
title=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    if pd.to_datetime(d.find(class_="date").text)==date.today():
        Date.append(d.find(class_="date").text)
        title.append(d.find_all('a')[-1].text)

df=pd.DataFrame({"時間":Date,"新聞標題":title})

```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
from datetime import date
import pandas as pd
import datetime

browser = webdriver.Chrome(ChromeDriverManager().install())

browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑

import time
for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

source=browser.page_source
soup=BeautifulSoup(source)

Date=[]
title=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (pd.to_datetime(datetime.datetime.now())-pd.to_datetime(d.find(class_="date").text)).seconds/3600 <=2:
        Date.append(d.find(class_="date").text)
        title.append(d.find_all('a')[-1].text)

df=pd.DataFrame({"時間":Date,"新聞標題":title})
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
from datetime import date
import pandas as pd
from selenium.webdriver.support.ui import Select


browser = webdriver.Chrome(ChromeDriverManager().install())

browser.get("https://www.ettoday.net/news/news-list.htm")

selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value('2020')
selectYear = Select(browser.find_element_by_id("selM"))
selectYear.select_by_value('3')
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value('18')
browser.find_element_by_id('button').click()

# 每個兩秒鐘自動往下滑

import time
for i in range(40):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 1000000);")

source=browser.page_source
soup=BeautifulSoup(source)

Date=[]
title=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    if pd.to_datetime(d.find(class_="date").text)>=pd.to_datetime("2020/03/18 15:00:00") and pd.to_datetime(d.find(class_="date").text)<=pd.to_datetime("2020/03/18 17:00:00"):
        Date.append(d.find(class_="date").text)
        title.append(d.find_all('a')[-1].text)

df=pd.DataFrame({"時間":Date,"新聞標題":title})
```
